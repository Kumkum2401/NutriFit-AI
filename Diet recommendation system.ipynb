```python
import os
import cohere 
import toml  # Library to read .toml files

# Load secrets.toml
secrets = toml.load(".streamlit/secrets.toml")

# Set API key from secrets
os.environ["COHERE_API_KEY"] = secrets["COHERE_API_KEY"]

# Initialize Cohere client using env variable
co = cohere.Client(os.environ["COHERE_API_KEY"])

generation_config = {
    "temperature": 0.6, 
    "max_tokens": 4096
}

# Input values 
name = "Sneha"
age = 21
gender = "Female"
weight = 60
height = 165
goal = "Weight Gain"
diet = "Vegetarian"
activity_level = "Moderate"
medical_conditions = "None"
region = "India"
state = "Uttar Pradesh"

# Construct the prompt
prompt = f"""Personalized Diet & Workout Plan:
Name: {name}
Age: {age}
Gender: {gender}
Weight: {weight} kg
Height: {height} cm
Goal: {goal}
Diet Preference: {diet}
Activity Level: {activity_level}
Medical Conditions: {medical_conditions}
Region: {region}
State: {state}
Provide a structured 7-day diet and workout plan."""

try:
    response = co.chat(
    message=prompt,
    model="command-r-plus",  # You can also use "command-r"
    temperature=0.6,
    max_tokens=2048
    )
    print(response.text)
except ValueError as e:
    print(f"‚ùå Error: {e}")
```
